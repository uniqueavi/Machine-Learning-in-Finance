```python
import pandas as pd
import numpy as np
import tensorflow as tf

from keras.layers import Dense
from keras.models import Sequential

from sklearn.metrics import confusion_matrix

from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
url = 'https://drive.google.com/file/d/1DsIOeH7z4WeOipjsQZuxJ3TPt-KAN-pH/view'
url = 'https://drive.google.com/uc?id=' + url.split('/')[-2]

print(url)
```

```python
dataset = pd.read_csv(url)

dataset
```

```python
dataset.columns
```

```python
for ind, col in enumerate(dataset.columns):
    print('{0} {1}'.format(ind, col))
```

```python
X = dataset.iloc[:, 3:13].values
y= dataset.iloc[:, 13].values
```

```python
lbl_encoder_1 = LabelEncoder()
X[:,1] = lbl_encoder_1.fit_transform(X[:,1])

lbl_encoder_2 = LabelEncoder()
X[:,2] = lbl_encoder_2.fit_transform(X[:,2])
```

```python
pd.DataFrame(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
len(X_train), len(X_test)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
pd.DataFrame(X_train)
```

```python
model = Sequential()
```

```python
model.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 10))

model.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))

model.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))

model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics =['accuracy'])
```

```python
model.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred = (y_pred > 0.5)
y_pred = pd.DataFrame(y_pred)
y_pred.describe()
```

```python
scores = model.evaluate(X_test, y_test)
```

```python
print('%s is: %.2f%%' % (model.metrics_names[1], scores[1]*100))
```

```python
model.metrics_names
```

```python
cm = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(cm)
cm_df.columns = ['Predicted No Exit', 'Predicted Exit']
cm_df.index = ['Actual No Exit', 'Actual Exit']
```

```python
# How to identify which label will be there for the confusion matrix
from sklearn.utils.multiclass import unique_labels

labels = unique_labels(y_test, y_pred)
print(labels)

#0 is for No Exit, 1 for Exit
```
